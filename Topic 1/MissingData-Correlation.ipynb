```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a skewed population (Gamma distribution)
np.random.seed(0)
population = np.random.gamma(shape=2, scale=2, size=50000)

# Take a sample from this skewed population
n = 40
sample = np.random.choice(population, size=n, replace=False)

# Bootstrap resampling
B = 5000
boot_means = [np.mean(np.random.choice(sample, size=n, replace=True)) for _ in range(B)]

# CLT approximation
clt_mean = np.mean(sample)
clt_sd = np.std(sample, ddof=1) / np.sqrt(n)
x = np.linspace(min(boot_means), max(boot_means), 500)
clt_pdf = 1/(clt_sd*np.sqrt(2*np.pi)) * np.exp(-(x - clt_mean)**2 / (2*clt_sd**2))

# Plot everything
plt.figure(figsize=(10, 6))
sns.histplot(boot_means, bins=40, color='skyblue', stat='density', label='Bootstrap Distribution')
plt.plot(x, clt_pdf, 'r-', label='CLT Normal Approximation')
plt.axvline(np.mean(sample), color='k', linestyle='--', label='Sample Mean')

plt.title("Bootstrap vs CLT Sampling Distribution of Sample Mean\n(Skewed Population)")
plt.legend()
plt.xlabel("Mean")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```
